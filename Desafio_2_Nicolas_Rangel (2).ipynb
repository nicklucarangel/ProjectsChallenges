```python
import pandas as pd
from pydataset import data
```

```python
df = data('mtcars')
df.reset_index(inplace=True)
df.rename(columns={'index':'cars'},inplace=True)
```

Quantos valores nulos existem em cada coluna?

```python
print(df.isnull().sum())
```

Quantos carros únicos existem? Retorne apenas o valor

```python
df['cars'].nunique()

#Viu, aqui vc não precisou printar já que existe o df.__str__ existe na classe

```

Qual é a média, o valor mínimo e valor mácimo de mpg(miles per gallon)

```python
# Média
Med_mpg = df['mpg'].mean()
print(Med_mpg)

#Top!
```

```python
# Mínimo
Min_mpg = df['mpg'].min()
print(Min_mpg)

#top!!
```

```python
# Máximo
Max_mpg = df['mpg'].max()
print(Max_mpg)

#Top!!
```

Crie um novo dataframe apenas com os carros de mpg acima ou igual a média.

```python
cars_between_med_max_mpg = df['mpg'].between(Med_mpg, Max_mpg, inclusive=True) #Setando condicional
print (cars_between_med_max_mpg) 
df[cars_between_med_max_mpg].sort_values(by=['mpg']) #Criando data frame e ordenando pela coluna "mpg"

#Achei bem interessante, e pode ser usado para fatiar de forma bem avançada, mas para resolver esse problema você poderia usar somente um 
df[df['mpg'] >= Med_mpg]
```
